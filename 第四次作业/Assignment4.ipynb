# Problem 2.1

```python
import xarray as xr
ds = xr.open_dataset("3B42_Daily.20181231.7.nc4", engine="netcdf4")
```

```python
da = ds.precipitation
```

```python
value = [];
lon = [];
lat = [];
for i in range(50):
    for x in range(10):
        value.append(da[20*i][20*x].values)
        lon.append(da[20*i][20*x].lon.values)
        lat.append(da[20*i][20*x].lat.values)

df = pd.DataFrame(list(zip(value, lon, lat)), columns =['precipitation', 'lon', 'lat'])
top = df.sort_values('precipitation', ascending=False).head(50)
top
```

```python
from matplotlib.lines import Line2D as Line
```

```python
# I got inspired by reading  https://scitools.org.uk/cartopy/docs/latest/
```

```python
# Create and define the size of a figure object 
plt.figure(figsize=(10,10), dpi=100)

# Create an axes with an basic PlateCarree projection style
proj = ccrs.PlateCarree()

ax = plt.axes(projection=proj)
plt.title("Top 50 precipitation of 2018.12.31")

ax.stock_img()
ax.coastlines()
# color map
cmap = plt.cm.Reds

# draw plot of earthquakes. 
cbar = plt.scatter(x=top['lon'], y=top['lat'], zorder=2,c=top['precipitation'], cmap=cmap, edgecolors='black')

plt.colorbar(cbar,ax=ax, label='precipitation', shrink=0.3)
ax.gridlines(crs=proj, linewidth=2, color='black',linestyle='--')

# Create a legend for the coastlines.
legend_artists = [Line([0], [0], color=color, linewidth=3)
                      for color in ('white', 'gray')]
legend_texts = ['Test 1', 'Test 2']
legend = ax.legend(legend_artists, legend_texts, fancybox=True,
                       loc='lower left', framealpha=0.75)
ax.set_global()

# masks or features
# I got inspired by reading https://stackoverflow.com/questions/36525825/mask-ocean-or-land-from-data-using-cartopy
ax.add_feature(cartopy.feature.LAND, zorder=10, edgecolor='k')
# annotations
ax.annotate('dense here', xy=(-150, -31), xytext=(-140, 5.5),
            arrowprops=dict(facecolor='red', shrink=0.05),
            )
# text box
ax.text(5, 5, 'Africa', color='red', size=15, ha='center', zorder=20, va='center', transform=proj)
# x label and ticks and y label and ticks
ax.set_xticks([-180, -120, -60, 0, 60, 120, 180], crs=proj)
ax.set_yticks([-78.5, -60, -25.5, 25.5, 60, 80], crs=proj)
ax.set_xlabel('lon')
ax.set_ylabel('lat')
plt.show()
```

# Problem 2.2 regional map

```python
ad = da.where(da['lon']>-40).where(da['lon']<20).where(da['lat']<40).where(da['lat']>-55)
```

```python
value = [];
lon = [];
lat = [];
for i in range(50):
    for x in range(10):
        value.append(ad[20*i][20*x].values)
        lon.append(ad[20*i][20*x].lon.values)
        lat.append(ad[20*i][20*x].lat.values)

df = pd.DataFrame(list(zip(value, lon, lat)), columns =['precipitation', 'lon', 'lat'])
top = df.sort_values('precipitation', ascending=False).head(50)
top
```

```python
# Create and define the size of a figure object 
plt.figure(figsize=(10,10), dpi=100)

# Create an axes with an basic PlateCarree projection style
proj = ccrs.PlateCarree()

ax = plt.axes(projection=proj)
plt.title("Top precipitation of 2018.12.31")

ax.stock_img()
ax.coastlines()
# color map
cmap = plt.cm.Reds

# draw plot of earthquakes. 
cbar = plt.scatter(x=top['lon'], y=top['lat'], zorder=2,c=top['precipitation'], cmap=cmap, edgecolors='black')

plt.colorbar(cbar,ax=ax, label='precipitation', shrink=0.3)
ax.gridlines(crs=proj, linewidth=2, color='black',linestyle='--')

# Create a legend for the coastlines.
legend_artists = [Line([0], [0], color=color, linewidth=3)
                      for color in ('white', 'gray')]
legend_texts = ['Test 1', 'Test 2']
legend = ax.legend(legend_artists, legend_texts, fancybox=True,
                       loc='lower left', framealpha=0.75)

ax.add_feature(cartopy.feature.LAND, zorder=10, edgecolor='k')
# annotations
ax.annotate('dense here', xy=(-20, -31), xytext=(-30, 5.5),
            arrowprops=dict(facecolor='red', shrink=0.05),
            )
# text box
ax.text(5, 5, 'Africa', color='red', size=15, ha='center', zorder=20, va='center', transform=proj)
# x label and ticks and y label and ticks
ax.set_xticks([-60,-40, -30,-20, -10,0, 10, 20,50], crs=proj)
ax.set_yticks([-60,-40, -25.5, 25.5,40,60], crs=proj)
ax.set_xlabel('lon')
ax.set_ylabel('lat')
ax.set_extent([-43, 20, -55, 30])

plt.show()
```
